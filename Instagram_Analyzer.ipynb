```python
import pandas as pd
import matplotlib.pyplot as plt
import re
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
file_path = 'c:/cr/펫푸드_20180205_2222.csv'
font_path_window="c:/Windows/Fonts/malgun.ttf"

#한글 파일을 읽을 때 오류가 날 수 있음, 그럴때 engine='python' 추가
data = pd.read_csv(file_path, engine='python')
```

```python
def tags():
    tags = []
    articles = data['본문']
    
    for article in articles:
        tags += re.findall(r'#[^\s#,\\]+', article)
    
    tag_count = Counter(tags)
      
    %matplotlib inline
    
    wordcloud = WordCloud(font_path= font_path_window, background_color="white", 
                          max_words=100).generate_from_frequencies(tag_count) #font_step=5
    plt.figure(figsize=(20,15))
    plt.imshow(wordcloud)
    plt.axis('off') 
    
    return tag_count
```

```python
a = tags()
```
